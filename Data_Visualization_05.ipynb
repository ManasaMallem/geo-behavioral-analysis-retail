```python
#
import matplotlib.pyplot as plt
import pandas as pd
# Example: Plot a bar chart of average spending by segment
df=pd.read_csv('Datasets/Store_segments.csv')
avg_spending = df.groupby('segment')['total_spent'].mean()
avg_spending.plot(kind='bar', color=['blue', 'green', 'orange'])
plt.title('Average Spending by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Spending')
plt.tight_layout()
plt.show()
```

```python
#spending by city(Bar chart)
city_spending = df.groupby('city')['total_spent'].sum().sort_values(ascending=False).head(10)
city_spending.plot(kind='bar', color='teal')
plt.title('Top 10 Cities by Total Spending')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Recency vs Monetary (Scatter Plot)
plt.figure(figsize=(8, 6))
plt.scatter(df['recency'], df['monetary'], c='purple', alpha=0.6)
plt.title('Recency vs Monetary Value')
plt.xlabel('Recency (days since last purchase)')
plt.ylabel('Total Spent')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#RFM Segment Distribution (Pie Chart)
df['segment'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'coral'])
plt.title('Customer Segment Distribution')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
